## Descriptive Analysis

```python
# Import Bibliotheken
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpl

# Einlesen der Daten in die Variable "data"
data = pd.read_csv("../data/zip.train", header=None, sep=" ")

# Ausgabe der Länge von data (Zeilen) 
print(len(data))
```

```python
# Zuweisen der ersten Zeile von data in Variable image
image = data.iloc[0, 1:257]

# Ausgabe der Länge von image
print("Länge von image: ", len(image))

# Umformung der Variablen image in eine 16x16 Struktur:
image_reshaped = image.values.reshape(16,16)
print('Länge von image-reshaped: ', len(image_reshaped))

#Shape von image_reshaped
image_reshaped.shape

```

```python
#Darstellung des ersten Datensatzes als Grafik:
mpl.imshow(image_reshaped, cmap="hot", interpolation="none")
mpl.show
```

```python
# Abschneiden der letzten Spalte
cleaned_data = data.iloc[:,:-1]
print("Shape cleaned data: ", cleaned_data.shape)

#Selektion aller Zeilen, die in der ersten Spalte den Wert 6 haben:
number_0 = pd.DataFrame(cleaned_data[cleaned_data[0] == 0])
number_1 = pd.DataFrame(cleaned_data[cleaned_data[0] == 1])
number_6 = pd.DataFrame(cleaned_data[cleaned_data[0] == 6])
number_7 = pd.DataFrame(cleaned_data[cleaned_data[0] == 7])
number_8 = pd.DataFrame(cleaned_data[cleaned_data[0] == 8])
number_9 = pd.DataFrame(cleaned_data[cleaned_data[0] == 9])
# print(number_6)
print("Länge Cleaned Data:", len(cleaned_data))
print("Länge Nummer 6:", len(number_6))

#Berechnung der Mittelwerte der Spalten über alle Zeilen bis zum vorletzten Wert
mean_image_6 = pd.DataFrame(number_6.iloc[:,:-1].mean())
print("Shape mean_image_6: ", mean_image_6.shape)

# print(image)
image = mean_image_6.values.reshape((16,16))
print("Image.shape:", image.shape)
mpl.imshow(image, cmap="hot", interpolation='nearest')
mpl.show()
```

```python
images = data.iloc[:,1:-1] # select all rows, columns 1 to end

fig=mpl.figure(figsize=(15, 15))
columns = 5
rows = 2
for i in range(0, columns*rows):
   digit_rows = images[data[0] == float(i)]
   mean_digit_rows = digit_rows.mean()
   img = mean_digit_rows.values.reshape(16,16)*0.5+0.5
   fig.add_subplot(rows, columns, i+1)
   mpl.imshow(img,cmap="jet",interpolation="nearest",vmin=0,vmax=1)
   mpl.colorbar(fraction=0.046, pad=0.04)
   mpl.tight_layout()
mpl.show()
```
