```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

Beispiel: Die Voraussage darüber, ob jeman einen Herzinfarkt bekommt, hängt ab von der Menge des zugeführten Cholesterins und der \n
    Eigenschaft Vegetarier. Nehmen wir weiterhin an, dass der Zusammenhang durch \
    ln(P(Y=1)/P(Y=0)) = 0.2*choles[0 - unendlich] -   veg[0 oder 1] beschrieben werden kann.
    
    

ln(P(Y=1)/P(Y=0)) = 2,5 #für choles = 10 und veg = 1

```python
# Erzeugen von Werten
choles = np.random.gamma(2,1,1000)
veg = np.round(np.random.uniform(0,1,1000))
noise = np.random.normal(0,2,1000)
```

```python
sample = pd.DataFrame()
sample['choles'] = choles
sample['veg'] = veg
sample['noise'] = noise
sample['logit'] = 0.2*choles - 2*veg #+ 0.5*noise
sample['p'] = np.exp(sample['logit'])/(np.exp(sample['logit'])+1)
sample['infarkt_y'] = np.round(sample['p'])
```

```python
sample.head()
```

## Regression:

```python
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression()

X = sample.iloc[:,0:2]
y = sample.iloc[:,5]

logReg.fit(X,y)

infarkt_predicted = logReg.predict(X)
sample['logReg'] = infarkt_predicted
```

```python
len(sample[sample['infarkt_y'] == 0])
```

```python
logReg.coef_
```

```python
print(len(sample[sample['infarkt_y'] == 1]))
print(len(sample[sample['logReg'] == 1]))
```

```python
logReg.score(X,y)
```
