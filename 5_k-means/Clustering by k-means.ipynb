```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x_values = np.random.normal(0,1,2500)
x_2_values = np.random.normal(10,2,2500)


#x_2_values = np.random.exponential(scale=2,size=1000)
y_values = np.random.normal(5,.5,5000)


#x_3_values = np.random.normal(10,12,1000)
#y_3_values = np.random.normal(10,12,1000)
```

```python
#Combine the two datasets
x_values= np.concatenate([x_values,x_2_values])
#data_x_2 = np.concatenate([x_2_values,y_2_values,y_2_values])
len(x_values)
```

```python
#Create a DataFrame to hold the combined data from x's and y's
```

```python
dataset = pd.DataFrame()
```

```python
dataset['x'] = x_values
dataset['y'] = y_values

#dataset.head(100)
```

```python
plt.figure(1,figsize=(10,10))
plt.scatter(dataset['x'] ,dataset['y'])
plt.show();
```

```python
kmeans = KMeans(n_clusters=4,random_state=0).fit(dataset)
dataset['cluster_nr'] = kmeans.predict(dataset)
```

```python
#Plot using scatter
plt.figure(1,figsize=(10,10))
plt.scatter(dataset['x'],dataset['y'],c=dataset['cluster_nr'])
plt.show();
```

```python
for i in range(0,4):
    print(len(dataset[dataset['cluster_nr'] == i]))
```
