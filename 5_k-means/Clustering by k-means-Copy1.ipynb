```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x_values = np.random.normal(0,1,50000)
#x_2_values = np.random.normal(10,5,2500)
#x_values= np.concatenate([x_values,x_2_values])
#x_2_values = np.random.exponential(scale=2,size=5000)

y_values = np.random.normal(-5,1,50000)
z_values = np.random.normal(5,1,50000)
f_values = x_values+y_values-z_values

#x_3_values = np.random.normal(10,12,1000)
#y_3_values = np.random.normal(10,12,1000)
```

```python
dataset = pd.DataFrame()
dataset['x'] = x_values
dataset['y'] = y_values
dataset['z'] = z_values
dataset['f'] = f_values
dataset.head()
```

```python
plt.figure(1,figsize=(10,10), )
plt.scatter(dataset['x'] ,dataset['f'])
plt.scatter(dataset['y'] ,dataset['f'])
plt.scatter(dataset['z'] ,dataset['f'])
plt.show();
```

```python
a=np.array(x_values)
b=np.array(y_values)
c=np.array(z_values)
d=np.array(f_values)
corr=np.corrcoef([np.array(x_values),b,c,d])
np.round(corr,3)
```

```python
kmeans = KMeans(n_clusters=3,random_state=0).fit(dataset)
dataset['cluster_nr'] = kmeans.predict(dataset)
```

```python
#Plot using scatter
plt.figure(1,figsize=(10,10))
plt.scatter(dataset['x'],dataset['y'], dataset['z'], c=dataset['cluster_nr'])
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);


plt.show();
```
